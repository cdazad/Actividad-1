# Manejo de repositorios
1.Crear nuevo repositorio en Github online
2. Colocar titulo 
3. seleccionar publico
4. selecionar Add REALME y crear
5. Adicionar archivos y colocar comentario de lo que se hizo en commit 
6. Va a code y copia la url del repositorio
EN LA CONSOLA DE GIT
7. git clone (pega url del repositorio)Enter
8. ls (Enter)
9. cd (nombre de la carpeta repositorio) 
10. ls (Enter) y trae nombre de los archivos adjuntos 
EDITAR ARCHIVOS DESDE LA CONSOLA 
11. vi README.md (editor) 
12. Agrega contenido (se pone lo que quiera agregar) 
13. Graba y sale (Esc :x)
14. git add . (esto es para Agregar modificaciòn)
15. git commit -m "se edito el readme"
16. git push (esto es para subir modificaciòn y se espera que actualice y ya se puede ir a cargar la version online y se ve reflejado el cambio)
